```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'../features_3_sec.csv')
```

```python
# Features of interest (means and variances)
df_means = df[["chroma_stft_mean", "rms_mean", "spectral_centroid_mean", "spectral_bandwidth_mean"]]
df_var = df[["chroma_stft_var", "rms_var", "spectral_centroid_var", "spectral_bandwidth_var"]]
```

```python
corr_means = df_means.corr()
sns.heatmap(corr_means, xticklabels=corr_means.columns, yticklabels=corr_means.columns, annot=True)
```

```python
corr_var = df_var.corr()
sns.heatmap(corr_var, xticklabels=corr_var.columns, yticklabels=corr_var.columns, annot=True)
```

## First Impression
It is interesting to see that spectral_centroid_mean and spectral_bandwidth_mean are highly correlated. I believe this implies that, on average, higher frequencies are correlated with a greater range of frequencies heard within a given time step. As someone who is mostly unfamiliar with these audio measurements, I am quite surprised. 

In terms of my own expectations, I expected rms_mean to be more highly correlated with spectral_centroid_mean. Perhaps this was just me drawing conclusions early that songs with higher frequencies should be louder on average. Instead, we have a weak correlation at best.

In terms of the variance chart, we notice that the variances between spectral centroids and bandwidths are indeed somewhat correlated. My conclusion from this is as follows: On average, the more the centroid moves, the more we expect the range of frequencies to change. This makes sense as more dynamic songs are expected to have wider ranges of frequencies. Oddly enough, the means for each of these metrics are more highly correlated than the variances.

```python
df_varmean = pd.concat([df_means, df_var], axis=1)
corr_varmean = df_varmean.corr()
sns.heatmap(corr_varmean, xticklabels=corr_varmean.columns, yticklabels=corr_varmean.columns, annot=True)
```
